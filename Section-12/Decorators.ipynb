## Decorators
- Simply provides way to modify or add new functionality to a function without changing the code of the function

```python
def main_welcome(func):
    greetings = "Hey!! There"
    def sub_welcome_method():
        print(greetings)
        print("Welcome to advance python course")

        func()
        print("Closure execution ends")
    return sub_welcome_method()

def course_intro():
    print('This is a Machine learing course')

main_welcome(course_intro)
```

**running the above without calling main_welcome**

```python
@main_welcome
def course_intro():
    print("This is a machine learning course")

course_intro
```

```python
def my_decorator(func):
    def wrapper():
        print("something before function")
        func()
        print("something after function")
    return wrapper()

@my_decorator
def say_hello():
    print("Hello World!!")
say_hello
```

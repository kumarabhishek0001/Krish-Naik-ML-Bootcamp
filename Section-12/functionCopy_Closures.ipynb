## Function Copy

```python
def welcome():
    return "Welcome to this Course"
welcome()
```

```python
welcome_copy = welcome

welcome_copy()
```

```python
print(welcome)
print(welcome_copy)

# welcome and welcome_copy are just names referencing to the same memory location
# python has grbage collector which frees up memory only when there are no name is bind to it.
```

```python
# even after deleting the main function we copied. We can still the use the function that was copied
del welcome

welcome_copy()
```

```python
welcome_copy()
```

## Closures

- A closure is a function that remembers the variables from the scope in which it was **created even after the scope has finished executing**

```python
# example  
def outer(name):
    greeting = f"Hello, {name}"

    def inner():
        print(greeting)
    return inner

# here we are returing a fucntion so,
# the return function has also to be called

say_hello = outer('Abhishek')
say_hello()
```

```python
print(say_hello.__closure__)
```

```python
# example
def main_welcome():
    msg = "Var from parent"
    def sub_welcome_method():
        print('accessing variable from the parent')
        print(msg)
        print('after accesing variable from parent')
    # in this return the function is not returned 
    # but the result of calling the funcion is reutrnd
    return sub_welcome_method()

main_welcome()
```

```python
# Example-2
def welcome(name):
    def say_hello():
        print(f"Hello, {name}")
        # we can access name inside closure
    return say_hello()
welcome('Abhishek')
```

```python
# function as parameter
def main_welcome(func):
    greetings = "Hello!! Welcome to this course"
    def sub_welcome_message():
        print("Something happening before calling the function")
        func(greetings)
        print("Something happening before calling the function")
    return sub_welcome_message()

main_welcome(print)
```

```python
# example-4
def welcome_message(func, lst):
    message = "Hello!!"

    def inner():
        print(message)
        print("Counting the number of items in the list")
        count = func(lst)
        print(f"number of items in list: {count}")
    return inner()

welcome_message(len,[1,2,3,4,5,6])

```

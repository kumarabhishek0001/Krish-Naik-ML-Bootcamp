# Generators

- Simple way to create iterators.
- They use yeild keyword to produce a series of values lazily.
- Meaning they generate values on the fly and do not store them into memory.

```python
# initializing
def square(n):
    for i in range(n):
        yield i**2
```

```python
# lazy loading
square(5)
```

```python
# Accessing elements 
for val in square(5):
    print(val)
```

```python
# accessing elements with changed
for val in square(3):
    print(val)
```

```python
# creating object from generator
nums = square(5)

# acessing elements inside the objects 
for val in nums:
    print(val)
```

```python
# using next to access elements
next(nums)
# next does not work here since the loop above calls the next to get the values
```

```python
# To be able to use next again initialize it again
nums = square(5)

next(nums)
```

### Practical usage
- to read large files because the allow you to process one line at a time without loading the entire file into memory

```python
def read_large_files(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line
```

```python
filepath = "./sample.txt"

file_data = read_large_files(filepath)

for line in file_data:
    print(line.strip())
```

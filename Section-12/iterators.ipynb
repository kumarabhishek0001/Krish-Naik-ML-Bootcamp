# Iterators

- Advaced concept in python used for efficient looping and memory management.
- Used to access elements of a collection sequenctially without exposing the underlying structure.

```python
# normal iteration
my_list = [1,2,3,4,5,6,7,8]
for i in my_list:
    print(i)
```

```python
# iterator
iterator = iter(my_list)

# type
print(type(iterator))

# uses lazyloading
print(iterator)
```

```python
next(iterator) #gives error when run after the 
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration as sr:
    print("Last element of the array crossed")
```

```python
# string iterator
word = "Hello"
iterator = iter(word)

print(next(iterator)) # Output: H
print(next(iterator)) # Output: e
```

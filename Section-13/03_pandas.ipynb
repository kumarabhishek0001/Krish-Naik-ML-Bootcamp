## Handling missing values

```python
df.isnull()
```

```python
# Tells if there are any null value in a column in true/false
df.isnull().any()
```

```python
# tells if any row has a missing value
df.isnull().any(axis=1)
```

```python
# total number of null values in a column
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
df
```

```python
# filling null values with mean in column sales
df['sales_fill_NA'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
# renaming columns
df.rename(columns={'Date':'Sales_Date'}).head()
```

```python
# Filling NA values with 0 and converting float -> int
df['Value'] = df['Value'].fillna(0).astype(int)
df.head()
```

```python
# apply function too elements in a column
df['new_values'] = df['Value'].apply(lambda x:x*2)
df.head()
```

```python
# data aggregation based on groups in a column

grouped_mean = df.groupby('Category')['Value'].mean()
```

```python
grouped_mean
```

```python
# making groupes based on differnt columns
grouped_mean1 = df.groupby(['Product', 'Region'])['Sales'].mean()
grouped_mean2 = df.groupby(['Product', 'Region'])['Value'].mean()
```

```python
print(grouped_mean1)
print(grouped_mean2)
```

```python
# Aggregate multiple function
grouped_aggregate = df.groupby(['Product','Region'])['Value'].agg(['mean', 'sum', 'count'])
grouped_aggregate
```

```python
# samples for merging dataframe
df1 = pd.DataFrame({'key':['a','b','c'], 'value':[1,2,3]})
df2 = pd.DataFrame({'key':['a','b','d'], 'value':[4,5,6]})
```

```python
merged_df1= pd.merge(df1,df2,on='key', how="inner")
merged_df1
```

```python
merged_df2= pd.merge(df1,df2,on='key', how="outer")
merged_df2
```

```python
merged_df3= pd.merge(df1,df2,on='key', how="left")
merged_df3
```

```python
merged_df4= pd.merge(df1,df2,on='key', how="right")
merged_df4
```

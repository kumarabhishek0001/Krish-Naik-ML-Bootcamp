# Seaborn
- Built on top of matplotlib

```python
# importing modules
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# loading inbuilt dataset 
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_df)
plt.title('Total Bills vs Tips')
```

```python
# Line Plot
sns.lineplot(x='size', y='total_bill', data=tips_df)
plt.title('Size vs Total Bill')
```

```python
# Can use categorical data 
# bar plot

sns.barplot(x='day', y='total_bill', data=tips_df, estimator=sum)
plt.title('day vs Total Bill')
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips_df)
plt.title('day vs total bill')
```

```python
# Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips_df)
plt.title('day vs total bill')
```

```python
sns.histplot(tips_df['total_bill'], bins=15, kde=True)
plt.title('Total Bills Histogram')
```

```python
# Kde Plot
sns.kdeplot(tips_df['total_bill'], fill=True)
plt.title("Bills Distribution")
```

```python
# Pair Plot
sns.pairplot(tips_df)
```

```python
corr = tips_df[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

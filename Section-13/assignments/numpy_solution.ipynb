```python
# 4
array = np.random.randint(1,17,(4,4))
array
```

```python
lst = []
print(lst)
```

```python
# row wise addition
for row_num, row in enumerate(array):
    sum = 0
    for col_num, col in enumerate(row):
        sum = sum + col
    lst.append(sum)

print(lst)
```

```python
array_shape = array.shape
col_size = array_shape[1]
col_sum_lst = []

# col wise addition
for i in range(0,col_size):
    data = array[:,i]
    print(data)
    sum = 0
    for j in data:
        sum = sum + j
    col_sum_lst.append(sum)

print(col_sum_lst)
```

```python
# Assignment 10
# masked array is used to handle invalid data
# a invalid data must not be accounted when working on dataset
# it skew the result
# masked elements are invalid 

import numpy.ma as ma
```

```python
array = np.random.randint(1,17,(5))
print(array)
```

```python
mx = ma.masked_greater(array, 10)
sum = mx.sum()
print(mx)
print(sum)
```

```python
array = np.random.randint(1,10,(3,3))
print(array)
```

```python
diagonal_elements = array.diagonal()
print(diagonal_elements)

masked_diagonal_array = ma.masked_values(array,diagonal_elements)
print(masked_diagonal_array)
```

```python
val = masked_diagonal_array.mean()
print(val)
```

```python
fixed_Array = masked_diagonal_array.filled(val)
print(fixed_Array)
```

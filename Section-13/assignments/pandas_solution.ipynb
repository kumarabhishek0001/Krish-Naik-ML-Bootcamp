```python
data = {
    'Category' : np.random.choice(['A','B','C'],size=10),
    'value' : np.random.randint(1,100,size=10)
}
df = pd.DataFrame(data)
print(df)
```

```python
sum_of_category = df.groupby('Category')['value'].sum()
print(sum_of_category)

mean_of_categories = df.groupby('Category')['value'].mean()
print(mean_of_categories)

count_of_categories = df.groupby('Category')['value'].count()
print(count_of_categories)
```

```python
df1 = pd.DataFrame({
    'A':np.random.randint(1,100,size=3),
    'B':np.random.randint(1,100,size=3)
})

df2 = pd.DataFrame({
    'C':np.random.randint(1,100,size=3),
    'D':np.random.randint(1,100,size=3)
})

print("Data 1: \n",df1)
print("Data 2: \n",df2)

# concatenate the DataFrames along the rows
concat_row = pd.concat([df1,df2],axis=0)
print("Concatenated DataFrame (rows): ")
print(concat_row)

# concatenate the dataframe along the columns
concat_col = pd.concat([df1,df2],axis=1)
print("Concatenated DataFrame (columns):")
print(concat_col)
```

```python
# Create two Pandas DataFrames with different columns
df1 = pd.DataFrame({'A': np.random.randint(1, 100, size=3), 'B': np.random.randint(1, 100, size=3)})
df2 = pd.DataFrame({'C': np.random.randint(1, 100, size=3), 'D': np.random.randint(1, 100, size=3)})
print("DataFrame 1:")
print(df1)
print("DataFrame 2:")
print(df2)

# Concatenate the DataFrames along the rows
concat_rows = pd.concat([df1, df2], axis=0)
print("Concatenated DataFrame (rows):")
print(concat_rows)

# Concatenate the DataFrames along the columns
concat_columns = pd.concat([df1, df2], axis=1)
print("Concatenated DataFrame (columns):")
print(concat_columns)
```

```python
# Assignment 6
date_rng = pd.date_range(start='2025-01-01',end='2025-12-31', freq='D')

df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0,100,size=(len(date_rng)))

df.set_index('date', inplace = True)
print("Original DataFrame:")
print(df)
```

```python
# resample dataframe to compute monthly mean
monthly_mean = df.resample('ME').mean()
print(monthly_mean)
```

```python
# Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31'
date_rng = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0, 100, size=(len(date_rng)))
df.set_index('date', inplace=True)
print("Original DataFrame:")
# print(df)

# Compute the rolling mean with a window of 7 days
rolling_mean = df.rolling(window=7).mean()
print("Rolling mean DataFrame:")
print(rolling_mean)
```

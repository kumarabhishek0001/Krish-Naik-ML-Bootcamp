```python
import sqlite3
```

```python
# Specialized Querying
# filtering based on department
def get_records_from_department(employee_department):
    conn = sqlite3.connect('test.db')
    curosr = conn.cursor()

    curosr.execute('SELECT * FROM employees WHERE department = ?',(employee_department,))
    records = curosr.fetchall()

    conn.close()
    for record in records:
        print(record)

get_records_from_department('HR')
```

```python
# updating data based on id 
def update_department(employee_id, new_department):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()

    cursor.execute('''
        UPDATE employees
            SET department = ?
            WHERE id = ?
    ''', (new_department, employee_id))
    conn.commit()
    conn.close()
    print('Employee data updated successfully')

update_department(2,'Engineering')
```

```python
# deteleting based on ID
def delete_employee_record(employee_id):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()

    cursor.execute('''
    DELETE FROM employees
        WHERE id = ?
    ''',(employee_id,))

    conn.commit()
    conn.close()

    print('Employee data deleted successfully')

delete_employee_record(8)
```

```python
# multiple important operations
# Assignment:8(2)
def add_department_foreign_key():
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()
    try:
        cursor.executescript('''
        -- sql does not allow certain modification with foreign key ON
        PRAGMA foreign_keys = off;
        
        -- will execute the script as one transcation if error is cauased the db can be rolled back
        BEGIN TRANSACTION;
        
        -- renaming employess
        ALTER TABLE employees RENAME TO old_employees;
                            
        -- creating table
        CREATE TABLE employees(
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL,
            age INTEGER,
            department TEXT,
            department_id INTEGER,
            
            -- creating and connecting foreign key
            FOREIGN KEY(department_id) REFERENCES departments(id)
        );
                            
        -- copying contents of table and pasting in new table
        INSERT INTO employees(id,name,age,department)
        SELECT id,name,age,department FROM old_employees;
        
        --  Deleting new table
        DROP TABLE old_employees;

        -- commiting changes
        COMMIT;
        -- turning on foreign key
        PRAGMA foreign_keys=on;
        ''')
        conn.commit()
        print('Table employees modified successfully')
    except Exception as e:
        conn.rollback()
        print('Error occured, db rolled back')
        print(e)
    finally:
        conn.close()
add_department_foreign_key()
```

```python
# Assignment:8(3)
def insert_data_into_departments_and_employees(department_id,department_name,employee_id,employee_name,employee_age,employee_department):
    conn = sqlite3.connect('test.db')
    cursor = conn.cursor()

    try:
        cursor.execute('''
            INSERT INTO departments(id,name)
            VALUES(?,?)
        ''',(department_id,department_name))
        cursor.execute('''
            INSERT INTO employees(id,name,age,department,department_id)
            VALUES(?,?,?,?,?)
        ''',(employee_id,employee_name,employee_age,employee_department,department_id))
        conn.commit()
        print("Department and Employee reocord inserted successfully")
    except Exception as e:
        conn.rollback()
        print("Error Occured, Transaction rolledback")
        print(e)
    finally:
        conn.close()

insert_data_into_departments_and_employees(1,"Engineering",12,"Robbin",22,"Engineering")
```

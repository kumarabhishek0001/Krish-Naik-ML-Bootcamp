# logging in python
It is a crucial aspect of any application providing a way to trace back events, errors and operational information. Python's built-in logging module offers flexible framework for emitting log message from python programs

```python
import logging

# configure the basic logging setting
# logging.basicConfig(level=logging.DEBUG)

# log message
logging.debug('This is debug message')
logging.info('this is a warning message')
logging.warning('This is warning message')
logging.error('This is error message')
logging.critical('This is a critical message')
```

```python
# configuring the logging
import logging
logging.basicConfig(
    filename = 'app.log',
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)

# log messages
logging.debug('This is debug message')
logging.info('this is a warning message')
logging.warning('This is warning message')
logging.error('This is error message')
logging.critical('This is a critical message')
```
